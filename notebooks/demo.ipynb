# PAYBACK Lightweight Assistant — Demo
Five distinct queries (different **languages** and **intents**) against the live Cloud Run service `https://payback-assistant-506488371374.europe-west3.run.app`, showing the raw JSON responses.

| # | Query | Language | Expected intent / action |
|---|---|---|---|
| 1 | Bitte zeige mir Angebote für günstige Windeln | de | search → recommend |
| 2 | I need stuff for a pasta dinner | en | discovery → theme basket |
| 3 | Suche Shampoo bei dm | de | navigational → route_to_partner |
| 4 | Was ist besser: Bio-Olivenöl oder normales Olivenöl? | de | comparison → compare |
| 5 | Ich habe ein Problem mit meinen PAYBACK Punkten | de | customer_support → handoff |

```python
import json, requests
BASE = 'https://payback-assistant-506488371374.europe-west3.run.app'
def ask(query, user_id='demo-notebook'):
    r = requests.post(f'{BASE}/assist', json={'query': query, 'user_id': user_id}, timeout=60)
    r.raise_for_status()
    print(json.dumps(r.json(), ensure_ascii=False, indent=2))
print('service:', requests.get(f'{BASE}/health', timeout=30).json())
```

## 1 — German · search (price-aware)

```python
ask('Bitte zeige mir Angebote für günstige Windeln')
```

## 2 — English · discovery (theme basket across partners)

```python
ask('I need stuff for a pasta dinner')
```

## 3 — German · navigational (partner-scoped route)

```python
ask('Suche Shampoo bei dm')
```

## 4 — German · comparison

```python
ask('Was ist besser: Bio-Olivenöl oder normales Olivenöl?')
```

## 5 — German · customer support (handoff, no products)

```python
ask('Ich habe ein Problem mit meinen PAYBACK Punkten')
```
